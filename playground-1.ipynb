```python
image_paths = []
for i in range(1, 525):
    image_paths.append("1-"+str(i).rjust(3, '0')+".dcm")
```

```python
data = []

for p in image_paths:
    # Create full image path
    image_path = IMAGE_PATH + p
    
    # Load image
    image_data = pydicom.dcmread(image_path)
    data.append(image_data.pixel_array)
```
